## Matpotlib

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

- простые линейные графики

```python
# 1-мерный массив (вектор строка)
x_array = np.arange(10)
x_array
```

```python
y = x_array + 10

# визуализация
plt.plot(x_array, y)
plt.show()
```

```python
y = x_array ** 2

# визуализация
plt.plot(x_array, y)
plt.show()
```

```python
y_1 = x_array ** 2
y_2 = x_array + 2 
y_3 = x_array + (x_array-10)


# визуализация
plt.plot(x_array, y_1)
plt.plot(x_array, y_2)
plt.plot(x_array, y_3)
plt.show()
```

```python
# 1-мерный массив из 100 значений
# линейная последовательность в диапазоне от 0 до 10
x = np.linspace(0, 10, 100)
x
```

```python
x = np.linspace(-10, 10, 1000)

# функция гиперболического тангенса
y = np.tanh(x)

# Функция сигмоиды
def sigmoid(x_arr):
    return 1 / (1 + np.exp(-x_arr))

y_2 = sigmoid(x)

# визуализация
plt.plot(x, y_1)
plt.plot(x, y_2)
plt.show()
```

```python
# цвет линий

y_1 = np.sin(x)
y_2 = np.cos(x)
y_3 = np.sin(x+1)
y_4 = np.sin(x+2)
y_5 = np.sin(x+3)
y_6 = np.sin(x+4)

plt.plot(x, y_1, color = "red") # по названию цвета
plt.plot(x, y_2, color = "m") # по первой букве цвета из форматов rgb cmyk
plt.plot(x, y_3, color = "#fa702a") #16-ричная кодировка rgb
plt.plot(x, y_4, color = (0.0, 0.8, 0.1)) # по формату RGB в виде кортежа со значениями от 0 до 1
plt.plot(x, y_5, color = (0.0, 0.1, 0.7))
plt.plot(x, y_6, color = "blue")

plt.show()
```

```python
# стиль линий
y_1 = np.sin(x)
y_2 = np.sin(x+1)
y_3 = np.sin(x+2)
y_4 = np.sin(x+3)

plt.plot(x, y_1, color = "red", linestyle = '-') # "solid" сплошная 
plt.plot(x, y_2,linestyle = '--') #'dashed' ) # штриховая линия
plt.plot(x, y_3,linestyle = ':') # "doted" пунктирная
plt.plot(x, y_4,linestyle = '-.' ) # штрих-пунктирная

plt.show()
```

```python
# Объединение настроек цвета и стиля линий

plt.plot(x, y_1, "g") 
plt.plot(x, y_2, "--r") 
plt.plot(x, y_3, ":b") 
plt.plot(x, y_4, "-.y")

plt.show()
```

```python
# Легенда графика

plt.plot(x, y_1, "g", label="синус") 
plt.plot(x, np.cos(x), "--r", label="косинус") 
plt.legend()

#диапазон координатной сетки
plt.axis([-15,15,-2,2])
plt.show()
```

```python
# Подпись осей

plt.plot(x, y_1, "g", label="синус") 
plt.plot(x, np.cos(x), "--r", label="косинус") 

plt.legend()
plt.xlabel("ось х")
plt.ylabel("sin and cos")

#диапазон координатной сетки
plt.axis([-15,15,-2,2])
plt.show()
```

```python
# Заголовок графика

# Размеры графика
plt.figure(figsize=(10,5))

plt.plot(x, y_1, "g", label="синус", marker="*") 
plt.plot(x, np.cos(x), "--r", label="косинус") 
plt.plot(x, np.cos(x+1), "or:", label="косинус")


plt.legend()
plt.xlabel("ось х")
plt.ylabel("sin and cos")
plt.title('Графики синуса и косинуса')
#диапазон координатной сетки
plt.axis([-15,15,-2,2])
plt.show()
```
